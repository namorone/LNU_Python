```python
def laguerre_func( t, n):
    beta = 2 
    sigma = 4
    #if t.all()<0 or n.all()<0 :
     #   print("Sorry, but t and n value has to be greater or equal than 0.")
      #  print('Your values: n = ', n , ';  t = ', t)
       # return
        
    lag0 = math.sqrt(sigma) * (math.e ** ((-1) * ((beta * t) / 2)))
    lag1 = lag0 * (1 - (sigma * t))

    if n == 0:
        return lag0
    elif n==1:
        return lag1
    elif n >= 2:
        for i in range(2, n+1):
            lag2 = ((2 * i - 1 - sigma * t)/i) * lag1 - ((i - 1)/i) * lag0
            lag0=lag1
            lag1=lag2
        return lag2
```

# Task 2


```python
def tabulate_laguerre( T, n):
    t=np.linspace(0, T, num=50)
    lag = [laguerre_func(i, n) for i in t]
    lag1 = [laguerre_func(i, n+1) for i in t]
    lag2 = [laguerre_func(i, n+2) for i in t]
    lag3 = [laguerre_func(i, n+3) for i in t]
    data = {
               'T': t,
                'L': [lag, lag1, lag2, lag3]
                }
    return data
```

```python
def construct_a_graph(data):
    plt.figure()
    plt.plot(data['T'],data['L'][0], 'violet', label = ' n')
    plt.plot(data['T'], data['L'][1], label = ' n+1')
    plt.plot(data['T'], data['L'][2], label = ' n+2')
    plt.plot(data['T'], data['L'][3], label = ' n+3')
    
    plt.grid(True)
    plt.title('LAGUERRE\'S FUNCTION')
    plt.xlabel('tabulation')
    plt.ylabel('laguerre results(t)')
    plt.legend()
    plt.show()
```

```python
construct_a_graph(tabulate_laguerre(20,5))
```

```python
def method_of_rectangles(f, a, b, num_of_points):
    n = ((b-a)/num_of_points)/2
    points = np.linspace(a+n, b-n, num_of_points)
    S = sum(f(points))*(b-a)/(num_of_points-1) 
    return S
```

```python
def inverse_transformation_laguerre (continuity, t ):
    lg = []
    for m in range(len(continuity)):
        lg.append(laguerre_func(t,m))
    
    h = np.sum(continuity * lg)
    return h
    
```
