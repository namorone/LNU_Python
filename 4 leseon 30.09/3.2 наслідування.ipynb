```python
#3.2 наслідування 

class Vehicle:
    def __init__(self, brand, engine_power, num_of_wheels, weight):
        self.brand = brand
        self.engine_power = engine_power
        self.num_of_wheels = num_of_wheels
        self.weight = weight
    
    
    
    @property
    def brand(self):
        return self._brand
    
    @property
    def engine_power(self):
        return self._engine_power
    
    @property
    def num_of_wheels(self):
        return self._num_of_wheels
    
    @property
    def weight(self):
        return self._weight
    
    @brand.setter
    def brand(self,value ):
        self._brand = value
    
    @engine_power.setter
    def engine_power(self,value ):
        self._engine_power = value
    
    @num_of_wheels.setter
    def num_of_wheels(self,value ):
        self._num_of_wheels = value
    
    @weight.setter
    def weight(self,value ):
        self._weight = value
    
    def __str__(self):
        return f'Brand: {self.brand} \t Power: {self.engine_power} \t Number of wheels: {self.num_of_wheels} \t Weight: {self.weight} \t'
    
class Truck(Vehicle) :
    def __init__(self, brand,  engine_power, num_of_wheels, weight, carrying_capacity):
        super().__init__(brand,  engine_power, num_of_wheels, weight)
        self.carrying_capacity = carrying_capacity
    
    def __str__(self):
        #super().__str__
        return f'Brand: {self.brand} \t Power: {self.engine_power} \t Number of wheels: {self.num_of_wheels} \t Weight: {self.weight} \t Carrying_capacity: {self.carrying_capacity} \t'
    
    @property
    def carrying_capacity(self):
        return self._carrying_capacity
    
    @carrying_capacity.setter
    def carrying_capacity(self,value ):
        self._carrying_capacity = value
    
class Car(Vehicle):
    def __init__(self, brand,  engine_power, num_of_wheels, weight, seats):
        super().__init__(brand,  engine_power, num_of_wheels, weight)
        self.seats = seats
    
    @property
    def seats(self):
        return self._seats
    
    @seats.setter
    def seats(self,value ):
        self._seats = value
        
    def __str__(self):
        #super().__str__
        return f'Brand: {self.brand} \t Power: {self.engine_power} \t Number of wheels: {self.num_of_wheels} \t Weight: {self.weight} \t Seats: {self.seats} \t'

class Bus(Car):
    def __init__(self, brand,  engine_power, num_of_wheels, weight, seats, standing):
        super().__init__(brand,  engine_power, num_of_wheels, weight, seats)
        self.standing = standing
    
    @property
    def standing(self):
        return self._standing
    
    @standing.setter
    def standing(self,value ):
        self._standing = value
        
    def __str__(self):
        #super().__str__
        return f'Brand: {self.brand} \t Power: {self.engine_power} \t Number of wheels: {self.num_of_wheels} \t Weight: {self.weight} \t Seats: {self.seats} \t Standing: {self.standing} \t'
```

```python
#оленки клас
class Vehicle:
    
    def __init__ (self, brand, power, wheel_num, weight):
        self.brand = brand
        self.power = power
        self.wheel_num = wheel_num
        self.weight = weight
        
    @property
    def brand(self):
        return self._brand
    
    @brand.setter
    def brand(self, value):
        self._brand = value
        
    @property
    def power(self):
        return self._power
    
    @power.setter
    def power(self, value):
        self._power = value
        
    @property
    def wheel_num(self):
        return self._wheel_num
    
    @wheel_num.setter
    def wheel_num(self, value):
        self._wheel_num = value
        
    @property
    def weight(self):
        return self._weight
    
    @weight.setter
    def weight(self, value):
        self._weight = value
    
    def __str__(self):
        return f'Brand: {self.brand} | Power: {self.power} | Number of wheels: {self.wheel_num} | Weight: {self.weight}'
        
```

```python
class Truck(Vehicle):
    
    def __init__(self, brand, power, wheel_num, weight, carrying_capacity):
        super().__init__(brand, power, wheel_num, weight)
        self.carrying_capacity = carrying_capacity
        
    def __str__(self):
        return super(Truck, self).__str__() + f' | Carrying_capacity: {self.carrying_capacity}'        
        
    @property
    def carrying_capacity(self):
        return self._carrying_capacity
    
    @carrying_capacity.setter
    def carrying_capacity(self, value):
        self._carrying_capacity = value
```

```python
class Car(Vehicle):
    
    def __init__ (self, brand, power, wheel_num, weight, sit_place):
        super().__init__(brand, power, wheel_num, weight)
        self.sit_place = sit_place
        
    def __str__(self):
        return super(Car, self).__str__() + f' | Number of sitting seats: {self.sit_place}' 
        
    @property
    def sit_place(self):
        return self._sit_place
    
    @sit_place.setter
    def sit_place(self, value):
        self._sit_place = value
        
    def total_number_of_seats(self):
        return self.sit_place
```

```python
class Bus(Car):
    
    def __init__ (self, brand, power, wheel_num, weight, sit_place, stand_place):
        super().__init__(brand, power, wheel_num, weight, sit_place)
        self.stand_place = stand_place
        
    def __str__(self):
        return super(Bus, self).__str__() + f' | Number of standing seats: {self.stand_place}' 
        
    @property
    def stand_place(self):
        return self._stand_place
    
    @stand_place.setter
    def stand_place(self, value):
        self._stand_place = value
        
    def total_number_of_seats(self):
        return self.sit_place + self.stand_place
```

```python
v1 = Vehicle('Toyota', 135, 4, 1600)
v2 = Vehicle('Hyundai', 108, 8, 2500)
v3 = Vehicle('Ford', 120, 6, 1800)

t1 = Truck('Hyundai', 170, 6, 5000, 3000)
t2 = Truck('Mercedes', 625, 10, 9202, 3000)
t3 = Truck('Renault', 380, 6, 9742, 9000)
t4 = Truck('Volkswagen', 28500, 10, 9000, 2500)

c1 = Car('Tesla', 503, 4, 2000, 7)
c2 = Car('Audi', 261, 4, 1545, 5)
c3 = Car('Opel', 103, 4, 1296, 5)
c4 = Car('Ferrari', 661, 4, 1475, 5)

b1 = Bus('Škoda', 150, 4, 11500, 27, 10)
b2 = Bus('BFM', 250, 4, 18000, 101, 0)
b3 = Bus('Ikarus', 326, 6, 15630, 28, 56)

autopark = [v1, v2, v3, t1, t2, t3, t4, c1, c2, c3, c4, b1, b2, b3]
```

```python
# 3 a
# повнi описи усiх транспортних засобiв автопарку

print('\t DESCIPTION OF ALL VEHICLES\n')
for i in autopark:
    print(i)
```

```python
# 3 b
# перелiк усiх транспортних засобiв автопарку, впорядкований за потужнiстю двигуна

autopark = sorted(autopark, key = lambda i: i.power)
print('\t A LIST OF ALL VEHICLES, SORTED BY ENGINE POWER\n')
for i in autopark:
    print(i)
```

```python
# 3 c
# окремi перелiки вантажiвок та пасажирських автомобiлiв, якi впорядкувати за спаданням вантажопiдйомнiстi 
# для вантажiвок i кiлькостi пасажирiв для пасажирських авто.

print('\t A LIST OF TRUCKS, REVERSE SORTED BY CARRYING CAPACITY\n')
trucks_list = [ i for i in autopark if (isinstance(i, Truck)) ]
trucks_list = sorted(trucks_list, key = lambda i: -i.carrying_capacity)
for i in trucks_list:
    print(i)
    
print('\n\t A LIST OF PASSENGER CARS, REVERSE SORTED BY THE NUMBER OF PASSENGERS\n')
cars_list = [ i for i in autopark if (isinstance(i, Car))]
cars_list = sorted(cars_list, key = lambda i: -i.total_number_of_seats())
for i in cars_list:
    print(i)
```
